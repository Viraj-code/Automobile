1 - Import Libraries

```python
import yfinance
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

2 - Download Data

```python
raw_data = yfinance.download(tickers="VOW3.DE, PAH3.DE, BMW.DE",
                            interval="1d",
                            group_by='ticker',
                            start="2009-01-29",
                            end="2022-06-01",
                            auto_adjust=True,
                            threads=True)
```

```python
raw_data.head()
```

```python
raw_data.tail()
```

3 - General Preprocessing

```python
df = pd.DataFrame([])
df['POR'] = raw_data['PAH3.DE'].Close
df['VW'] = raw_data['VOW3.DE'].Close
df['BMW'] = raw_data['BMW.DE'].Close
df['Vol_POR'] = raw_data['PAH3.DE'].Volume
df['Vol_VW'] = raw_data['VOW3.DE'].Volume
df['Vol_BMW'] = raw_data['BMW.DE'].Volume
df.head()
```

```python
# Find out number of missing values for each Stock Exchanges
df.isna().sum()
```

```python
# Setting the frequency of the data
# h = hourly, w = Weekly, d = daily, m = monthly, b = business days
df=df.asfreq('b')
```

```python
df.isna().sum()
```

```python
df=df.fillna(method='ffill') # Front-Filling any missing value
df = df.fillna(method='bfill') # Back-Filling any missing value
df.head()
```

```python
# Displaying the first and last 5 elements to make sure the data was scrapped correctly
print(df.head())
print(df.tail())
```

4 - Export Database

```python
df.to_csv("automobile-indexes.csv")
```
